## Plotting a simple line graph

```python
# import pyplot module using alias plt
# (standard convention is to use plt for pyplot)
import matplotlib.pyplot as plt

# data to plot
squares = [1, 4, 9, 16, 25]

# follow another common convention: call subplots() function
# this function can generate one or more plots in the same figure
# the variable fig represents the entire figure or collection of plots
# the variable ax represents a single plot in the figure
# (ax is the variable we'll use most of the time)
fig, ax = plt.subplots()
# use the plot() method to plot the data it's given in a meaningful way
ax.plot(squares)

# plt.show() opens matplotlib's viewer and displays the plot
plt.show()
```

### Changing the label type and line thickness

```python
# import
import matplotlib.pyplot as plt

# data to plot
squares = [1, 4, 9, 16, 25]

# generate plot
fig, ax = plt.subplots()
ax.plot(squares, linewidth = 3) # line thickness

# set chart title and label axes
ax.set_title("Square numbers", fontsize = 24)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize = 14)

# set size of tick labels
ax.tick_params(axis = 'both', labelsize = 14)

# show plot
plt.show()
```

### Plotting and styling individual points with `scatter()`

Plot a single point with the `scatter()` method

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(2, 4, s = 200)

# set chart title and label axes
ax.set_title("Square numbers", fontsize = 24)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize = 14)

# set size of tick labels
ax.tick_params(axis = 'both', which = 'major', labelsize = 14)

plt.show()
```

### Plotting a series of points with `scatter()`

```python
import matplotlib.pyplot as plt

# data to plot
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

# generate plots
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s = 100)

# set chart title and label axes
ax.set_title("Square numbers", fontsize = 24)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize = 14)

# set size of tick labels
ax.tick_params(axis = 'both', which = 'major', labelsize = 14)

plt.show()
```

### Calculating data automatically
Plotting 1000 points

```python
# import
import matplotlib.pyplot as plt

# data to visualize
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

# generate scatterplot
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s = 10)

# set chart title and label axes
ax.set_title("Square numbers", fontsize = 24)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize = 14)

# set range for each axis
ax.axis([0, 1100, 0, 1100000])

# show plot
plt.show()
```

### Defining custom colors
Pass `c` to `scatter()` with the name of a color to use in quotes.

```python
# import
import matplotlib.pyplot as plt

# data to visualize
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

# generate scatterplot
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s = 10, c = 'purple')

# set chart title and label axes
ax.set_title("Square numbers", fontsize = 24)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize = 14)

# set range for each axis
ax.axis([0, 1100, 0, 1100000])

# show plot
plt.show()
```

Can also pass custom colors with the RGB color model.

```python
# import
import matplotlib.pyplot as plt

# data to visualize
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

# generate scatterplot
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s = 10, color = (0, 0.8, 0))

# set chart title and label axes
ax.set_title("Square numbers", fontsize = 24)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize = 14)

# set range for each axis
ax.axis([0, 1100, 0, 1100000])

# show plot
plt.show()
```

### Using a colormap
A *colormap* is a series of colors in a gradient that moves from a starting to an ending color.  
To use a colormap, you need to specify how pyplot should assign a color to each point in the data set.
* Pass the data, `y_values` in this example, to `c`.
* Tell pyplot which colormap to use using the `cmap` argument.

```python
# import
import matplotlib.pyplot as plt

# data to visualize
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

# generate scatterplot
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s = 10, c = y_values, cmap = plt.cm.Blues)

# set chart title and label axes
ax.set_title("Square numbers", fontsize = 24)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize = 14)

# set range for each axis
ax.axis([0, 1100, 0, 1100000])

# show plot
plt.show()
```

### Saving plots automatically
Replace the `plt.show()` call with `plt.savefig()`.
* The first argument is the filename.
* The second argument trims extra whitespace around the plot.

```python
# import
import matplotlib.pyplot as plt

# data to visualize
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

# generate scatterplot
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s = 10, c = y_values, cmap = plt.cm.Blues)

# set chart title and label axes
ax.set_title("Square numbers", fontsize = 24)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Square of Value", fontsize = 14)

# set range for each axis
ax.axis([0, 1100, 0, 1100000])

# show plot
plt.savefig('squares_plot.png', bbox_inches = 'tight')
```
